```python
import requests
import urllib.parse
```

```python
def get_lat_long(Location):
    
    print(Location)
    
    url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(Location) +'?format=json'
    response = requests.get(url).json()
    x = (response[0]["lat"])
    y = (response[0]["lon"])
    
    return x,y
```

```python
destination_city['Destination_Log_Lat'] = destination_city['Destination_city'].apply(lambda x: get_lat_long(x))
```

```python
from math import sin, cos, sqrt, atan2, radians
def get_km(l1, lo1, l2, lo2):
    
    print(l1)
    # approximate radius of earth in km
    print(type(l1))
    l1 = float(l1)
    print(type(l1))
    lo1 = float(lo1)
    l2 = float(l2)
    lo2 = float(lo2)
    
    R = 6373.0

    lat1 = radians(l1)
    lon1 = radians(lo1)
    lat2 = radians(l2)
    lon2 = radians(lo2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    distance = R * c+280

    print("Result:", distance)
    return distance
```

```python
big_df['Combined_Location_km']= big_df['Combined_Location'].apply(lambda x: get_km(x.split("'")[1],x.split("'")[3],x.split("'")[5],x.split("'")[7]))
```
